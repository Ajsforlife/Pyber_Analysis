```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.bar(x_axis,y_axis)
```

```python
#create the plot
plt.bar(x_axis,y_axis,color="Green",label='Boston')
#create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('fare($)')
#create a title
plt.title("PrBer Fare by Month")
#add the legend
plt.legend()
```

```python
#create horizontal bar chart
plt.barh(x_axis,y_axis)
```

```python
#create horizontal bar chart with flipped axes
plt.barh(y_axis,x_axis)
```

```python
#create the plot flipping axis order
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
#Skill drill

#create the plot
plt.barh(x_axis,y_axis,color="Magenta",label='Boston')
#create labels for the x and y axes
plt.xlabel('Fare($)')
plt.ylabel('Date')
#create a title
plt.title("PrBer Fare by Month")
#add the legend
plt.legend()
#invert y-axis
plt.gca().invert_yaxis()
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
#create the plot with ax.plt() horizontal
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis)
```

```python
#switch axes in plot
fig, ax = plt.subplots()
ax.barh(y_axis,x_axis)
```

```python
#Skill drill

#create the plot
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color="Cyan",label='Chicago')
#create labels for the x and y axes
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#create a title
ax.set_title("PrBer Fare by Month")
#add the legend
ax.legend()
#invert y-axis
ax.invert_yaxis()
```
