```python
import matplotlib.pyplot as plt
```

```python
x = [50,42,39,34,32,48,22,17,8,13]
y = [17,19,21,22,24,23,34,31,43,35]
```

```python
plt.scatter(x,y,c="green",marker="x",s=x)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Average Fare vs Total Rides")
plt.show
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create a scatter plot plt.plot()
plt.plot(x_axis,y_axis,'o')
```

```python
#create a scatter plot plt.scatter()
plt.scatter(x_axis,y_axis)
```

```python
#skill drill
#create plot
plt.plot(y_axis,x_axis,'o',c="red",label='Chicago')
#label axes
plt.xlabel("Fare ($)")
plt.ylabel("Date")
#set x limit
plt.xlim(0,50)
#add title
plt.title("PyBer Fare by Month")
#add legend
plt.legend()
#invert y axis
plt.gca().invert_yaxis()
plt.show
```

```python
#create a bubble chart(3 or 4 factors)
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
#change size of the bubbles
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
#change size based on list comprehension
#new_list - [expression for item in list if conditional]
plt.scatter(x_axis,y_axis, s = [i *3 for i in y_axis])
```

```python
#create scatter plot using ax.scatter
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#create bubble plot using ax.scatter
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis,s=y_axis)
```

```python
#skill drill

#create bubble plot using ax.scatter
fig, ax = plt.subplots()
#change color to skyblue size to 5x, 20% transparceny, black line around circles, linewidth 2pts, add label
ax.scatter(y_axis,x_axis,c="skyblue",s=[i *5 for i in y_axis],alpha = .8,edgecolors='black',linewidth=2,label='Boston')
#add legend
ax.legend()
#add title
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
#add limit to x axis
ax.set_xlim(0,50)
#invert y axis
ax.invert_yaxis()
```
